### やること
* データ点 $\{x_1, ... , x_{N}\} \in \{1, 0\}$ を作成
* この$\vec{X}$がベルヌーイ分布から生成されているとしてパラメータ $\mu$ を推定する
* $\mu$ を推定するために，ベータ分布を用いる
    * 本来 $a, b$ はハイパーパラメータ

```julia
# μ
mu = 0.25

# range
r = range(0, stop=1, step=0.005)

# 適当な初期値
a = 0.5
b = 0.5

# データ数
N = 90

# データ作成
d = Bernoulli(mu)
X = rand(d, N)
println("X:", X)
println("mean:", mean(X))

# N=10
d = Beta(a, b)
y = pdf.(d, r)
plot(r, y)

x_sum = sum(X[1:10])

a_hat = x_sum + a
b_hat = N - x_sum + b

d = Beta(a_hat, b_hat)
y = pdf.(d, r)
plot!(r, y)

# N=30
x_sum = sum(X[1:30])
a_hat = x_sum + a
b_hat = N - x_sum + b
d = Beta(a_hat, b_hat)
y = pdf.(d, r)
plot!(r, y)


# N=50
x_sum = sum(X[1:50])
a_hat = x_sum + a
b_hat = N - x_sum + b
d = Beta(a_hat, b_hat)
y = pdf.(d, r)
plot!(r, y)

# N=70
x_sum = sum(X[1:70])
a_hat = x_sum + a
b_hat = N - x_sum + b
d = Beta(a_hat, b_hat)
y = pdf.(d, r)
plot!(r, y)

# N=90
x_sum = sum(X[1:90])
a_hat = x_sum + a
b_hat = N - x_sum + b
d = Beta(a_hat, b_hat)
y = pdf.(d, r)
plot!(r, y)
```

### 未観測データの分布予測

```julia
mu_hat = a_hat / (a_hat + b_hat)
d = Bernoulli(mu_hat)
X_unsight = rand(d, N)
println("X:", X_unsight)
println("mean:", mean(X_unsight))
```
